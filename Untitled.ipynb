```python
import numpy as np
import scipy as sp
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import glob

df1 = pd.read_csv('Time Band Pass Aluminum Reflection.S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
df2 = pd.read_csv('Aluminum Reflection at 4 inches without gating.S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
df3 = pd.read_csv('Aluminum Reflection at 4 inches with gating.S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
```

```python
fig = plt.figure(figsize=(24,12))

ax1 = fig.add_subplot(111, xlabel = 'Freq[Ghz]', ylabel = 'Power', title = 'Time Band Pass Example')

ax1.plot(df1.Freq, df1.Power**2, color= 'darkred')

plt.axis([0,100,0,1])

plt.show()
```

```python
fig = plt.figure(figsize=(12,6))

darkred_patch = mpatches.Patch(color='darkred', label='No Time Gating')
blue_patch = mpatches.Patch(color='blue',label='Time Gating')

ax1 = fig.add_subplot(111, xlabel = 'Freq[Ghz]', ylabel = 'Power', title = 'Gating Comparison')

ax1.plot(df2.Freq, df2.Power**2, color= 'darkred')
ax1.plot(df3.Freq, df3.Power**2, color= 'blue')

plt.axis([75,110,0,.2])

plt.legend(handles=[darkred_patch, blue_patch])

plt.savefig('Gating Comparison'+'.pdf', bboxes_inches= 'tight')
plt.show()
```
