```python
import numpy as np
import scipy as sp
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import glob

dfR1 = pd.read_csv('circular_horn_no_gating_open_room.S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfR2 = pd.read_csv('circular_horn_gating_aluminum(3.49cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfR3 = pd.read_csv('circular_horn_gating_aluminum(6.03cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfR4 = pd.read_csv('circular_horn_gating_aluminum(8.57cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfR5 = pd.read_csv('circular_horn_gating_aluminum(11.11cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfC1 = pd.read_csv('rectangular_horn_no_gating_open_room.S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfC2 = pd.read_csv('rectangular_horn_gating_aluminum(3.49cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfC3 = pd.read_csv('rectangular_horn_gating_aluminum(6.03cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfC4 = pd.read_csv('rectagular_horn_gating_aluminum(8.57cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
dfC5 = pd.read_csv('rectagular_horn_gating_aluminum(11.11cm).S1P', index_col=None, skiprows=14, delim_whitespace=True, header=None, names = ['Freq', 'Power', 'Phase'])
```

```python
fig = plt.figure(figsize=(24,12))

darkgreen = mpatches.Patch(color='darkgreen',label='Rectangular Horn Aluminum at 3.49cm')
green = mpatches.Patch(color='g', label='Rectangular Horn at 6.03cm')
mediumgreen = mpatches.Patch(color='mediumseagreen',label='Rectangular Horn AL at 8.57cm')
lightgreen = mpatches.Patch(color='lightgreen',label='Rectangular Horn AL at 11.11cm')

darkred = mpatches.Patch(color='darkred',label='Circular Horn Aluminum at 3.49cm')
red = mpatches.Patch(color='red',label='Circular Horn AL at 6.03cm')
lightred = mpatches.Patch(color='tomato',label='Circular Horn AL at 8.57cm')
lightsalmon = mpatches.Patch(color='lightsalmon',label='Circular Horn AL at 11.11cm')


ax1 = fig.add_subplot(111, xlabel = 'Freq[Ghz]', ylabel = 'Power', title = 'Aluminum Reflection with Circular and Rectangular Horns')

ax1.plot(dfR2.Freq, dfR2.Power**2, color= 'darkred')
ax1.plot(dfR3.Freq, dfR3.Power**2, color= 'red')
ax1.plot(dfR4.Freq, dfR4.Power**2, color= 'tomato')
ax1.plot(dfR5.Freq, dfR5.Power**2, color= 'lightsalmon')


ax1.plot(dfC2.Freq, dfC2.Power**2, color= 'darkgreen')
ax1.plot(dfC3.Freq, dfC3.Power**2, color= 'g')
ax1.plot(dfC4.Freq, dfC4.Power**2, color= 'mediumseagreen')
ax1.plot(dfC5.Freq, dfC5.Power**2, color= 'lightgreen')

plt.axis([75,110,0,.15])
plt.legend(handles=[darkgreen, green, mediumgreen, lightgreen, darkred, red, lightred, lightsalmon])

plt.savefig('Horn Comparison1'+'.pdf', bboxes_inches= 'tight')
plt.show()
```

```python
fig = plt.figure(figsize=(24,12))

green_patch = mpatches.Patch(color='purple', label='Rectangular Horn')
red_patch = mpatches.Patch(color='blue',label='Circular Horn')

ax1 = fig.add_subplot(111, xlabel = 'Freq[Ghz]', ylabel = 'Power', title = 'Horns with Open Room and No Gating')
ax1.plot(dfR1.Freq, dfR1.Power**2, color= 'blue')

ax1.plot(dfC1.Freq, dfC1.Power**2, color= 'purple')

plt.axis([75,110,0,.2])
plt.legend(handles=[red_patch, green_patch])

plt.savefig('Horn Comparison2'+'.pdf', bboxes_inches= 'tight')
plt.show()
```
